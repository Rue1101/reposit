NumPy — это открытая бесплатная Python-библиотека для работы с многомерными массивами, этакий питонячий аналог Matlab. NumPy чаще всего используют в анализе данных и обучении нейронных сетей — в каждой из этих областей нужно проводить много вычислений с такими матрицами.

```python
#a Импортируйте библиотеку NumPy
import numpy as np
```

```python
#b Создайте массив из 10 нулей
a = np.zeros(10)
print(a)
```

```python
#c Создать массив из 10 единиц
b = np.ones(10)
print(b)
```

```python
#d Создать массив из 10 пятерок
c = b*5
print(c)
```

```python
#e Создать массив из целых чисел от 10 до 50
d = np.arange(10, 100) #начало конец шаг
print(d)
```

```python
#f Создать массив из четных чисел от 10 до 50
f = np.arange(10, 51, 2)
print(f)
```

```python
#g Создать матрицу 3х3 с числами от 0 до 8
Z = np.arange(9).reshape(3,3)
print(Z)
```

```python
#h Создать единичную матрицу 3x3
h = np.eye(3)
print(h)
```

```python
#i Используйте NumPy для генерации случайного числа от 0 до 1
i = np.random.rand(1)
print(i)
```

```python
#j Используйте NumPy для генерации массива из 25 случайных чисел
j = np.random.rand(25)
print(j)
```

```python
#k Создайте следующую матрицу используя функции библиотеки NumPy 
k = np.arange(0, 1.01, 0.01)
print(k)
```

```python
# Матрица из условия задачи
mat = np.arange(1,26).reshape(5,5)
print(mat)
```

```python
#l Необходимо написать код который вернет следующую матрицу из стартовой:
print('l =', mat[2:, 1:])
```

```python
#m Написать код обращения к элементу который вернет значение  20 из стартовой матрицы задачи l.
ar = mat[mat == 20]
print(ar)
```

```python
#n Написать код к стартовой матрице задачи l , который воспроизводит вывод, показанный ниже
print('n =', mat[0:3, 1].reshape(3,1))
```

```python
#o Написать код к стартовой матрице задачи l , который воспроизводитвывод, показанный ниже 
print('o =', mat[4])
```

```python
#p Написать код к стартовой матрице задачи l , который воспроизводитвывод, показанный ниже
print('p =', mat[3:])
```

```python
#q Получить сумму всех значений в стартовой матрице mat
print(mat.sum())
```

```python
#r Получить стандартное отклонение в стартовой матрице mat
print(mat.std())
```

```python
#s Получить сумму значений в колонках в стартовой матрице mat 
print(mat.sum(axis=0))
```
